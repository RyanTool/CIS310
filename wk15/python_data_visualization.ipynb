```python
d = {'col1':[1,2,3,4,2,1,6,7,3,4], 'col2':[5,8,9,10,12,8,6,2,8,7]}
df2 = pd.DataFrame(d)
df2
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.plot(df2['col1'],'k.', label='col1')
ax.plot(df2['col2'],'k--', label='col2')
ax.legend(loc='best')
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.plot(df2['col1'],'ko', label='col1')
ax.plot(df2['col2'],'ko--', label='col2')
ax.legend(loc='best')
```

```python
#Ticks, Labels and Legends
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(randn(1000).cumsum())
```

```python
ticks = ax.set_xticks([0, 250, 500, 750, 1000])

labels = ax.set_xticklabels(['one','two','three','four','five'],rotation=30, fontsize='small')

ax.set_title('My First Matplotlib Plot With Ticks')

ax.set_xlabel('Stages')

ax.set_ylabel('Values')

fig
```

```python
#Send as dictionary through set function instead of doing it all manually
props = {'title': 'My First Matplotlib Plot With Ticks',
        'xlabel': 'Stages',
        'ylabel': 'Values'}

ax.set(**props)
fig
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.plot(randn(1000).cumsum(), 'k', label='One')
ax.plot(randn(1000).cumsum(), 'k--', label='Two')
ax.plot(randn(1000).cumsum(), 'k.', label='Three')

ax.legend(loc='best')
```
